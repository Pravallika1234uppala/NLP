# STEMMING:

## 1. Porter Stemmer:

```python
from nltk.stem import PorterStemmer

porter_stemming = PorterStemmer()

#applying porter stemming for each word:

for word in words:
    print(word+" ----> "+porter_stemming.stem(word))
```

## 2. Regular expression stemmer ( RegexpStemmer)

```python
from nltk.stem import RegexpStemmer
#it takes regular expression and removes prefix and suffix that matches.
reg_stemmer = RegexpStemmer('ing$|s$|tion$|ary$|able$', min=4)

#applying Regular expression stemming for each word:

for word in words_punct:
    print(word+" ----> "+reg_stemmer.stem(word))
```

## 3. Snowball Stemmer:

```python
from nltk.stem import SnowballStemmer

#function intialization:
snowball_stem = SnowballStemmer('english')

#applying snowball expression stemming for each word:

for word in words_punct:
    print(word+" ----> "+snowball_stem.stem(word))
```

# TEXT PROCESSING ----- STEP 2 -----> CONVERTING TEXT TO VECTORS:

```python
import pandas as pd
messages = pd.read_csv('email.csv', names=["label", "message"])
```

```python
messages=messages.drop(0)
```

```python
messages
```

##### Data Cleaning & preprocessing:

```python
import re
import nltk

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

#Intializing stemmer function:
ps = PorterStemmer()

#List to save cleaned data
corpus=[]

for i in range(1, len(messages)):
    review = re.sub('[^a-zA-z]', ' ', messages['message'][i])
    review.lower()
    #list conversion
    review=review.split()
    #stemming:
    review = [ps.stem(wrd) for wrd in review if wrd not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#Doing lemmatization for more accuaracy:
#Intializing lemmatizer function:
lemm = WordNetLemmatizer()

#List to save cleaned data
corpus=[]

for i in range(1, len(messages)):
    review = re.sub('[^a-zA-z]', ' ', messages['message'][i])
    review.lower()
    #list conversion
    review=review.split()
    #lemmatization:
    review = [lemm.lemmatize(wrd, pos='v') for wrd in review if wrd not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```
